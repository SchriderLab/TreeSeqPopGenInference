```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
preds = pd.read_csv("preds.csv")
names = ["N0", "t1", "N1", "t2", "N2"]
```

```python
for name in names:
    plt.scatter(preds[f"pred_{name}"], preds[f"true_{name}"])
    plt.title(name)
    plt.ylabel("True")
    plt.xlabel("Pred")
    plt.xscale("log")
    plt.yscale("log")
    plt.tight_layout()
    plt.show()
```
